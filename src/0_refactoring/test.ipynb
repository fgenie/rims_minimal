```python
import jsonlines
with jsonlines.open('/data/recoteam_583/joel/opensource/rims_minimal/dataset/gsm8K_test.jsonl') as f:
    records = list(f)
```

```python
import pandas as pd

def dedup(records):
    df = pd.DataFrame(records)
    k = "question" if "question" in df.columns else "problem"
    df = df.drop_duplicates(subset=k)
    records = df.to_dict(orient="records")
    return records

records = dedup(records)
```
